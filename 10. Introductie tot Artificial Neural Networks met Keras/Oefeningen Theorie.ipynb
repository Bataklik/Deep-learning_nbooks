3. (Oefening 6 uit het boek) Veronderstel dat een MLP bestaat uit <ins>n invoerlaag</ins> met __10 neuronen__, een verborgen laag met __50 neuronen__ en een uitvoerlaag met __3 neuronen__. Alle neuronen in de verborgen en de uitvoerlaag gebruiken de ReLU-activatiefunctie.
   - Wat is de vorm (m.a.w. het aantal rijen en kolommen) van de invoer-matrix $X$?
   - Wat is de vorm van de gewichten matrix $W_h$ die de gewichten van de verborgen laag bevat? Wat is de vorm van de biasvector $b_h$?
   - Wat is de vorm van de gewichtenmatrix $W_o$ die de gewichten van de uitvoerlaag bevat? Wat is de vorm van de biasvector $b_o$?
   - Wat is de vorm van de uitvoer $Y$ van het netwerk?
   - Geef de wiskundige vergelijking die $Y$ uitdrukt als functie van $X$, $W_h$, $b_h$, $W_o$ en $b_o$

```python
import keras


inputs = keras.Input(shape=(10,))

hidden = keras.layers.Dense(50, activation="relu")(inputs)

output = keras.layers.Dense(3, activation="relu")(hidden)

model = keras.Model(inputs=inputs, outputs=output)

print(model.summary())
```
